```python
import matplotlib.pyplot as plt
import seaborn as sns
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
x=titanic.survived.value_counts()[1]
percent =(x/891)*100
x=100-percent
data=[percent,x]
lable1=['survived','not survived']
plt.pie(data,labels=lable1,autopct="%1.1f")
plt.show()
```

```python
x=titanic.sex
sns.histplot(x)
```

```python
x=titanic.age
sns.histplot(x)
```

```python
sns.barplot(x="sex",y="survived",hue="pclass",data=titanic)
```

```python
class_survival = titanic.groupby('pclass')['survived'].mean().reset_index()
sns.barplot(x='pclass', y='survived', data=class_survival)
plt.title("Survival Rate by Journey Class")
plt.xlabel("Journey Class")
plt.ylabel("Survival Rate")
```
