```python
#Exploratory Data Analysis
import pandas as pd
import seaborn as sns
dataset=pd.read_csv("E:\data science\dataset\car data.csv")
dataset.describe()
```

```python
#preprocessing
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
fuel_type=dataset[['Fuel_Type']]
dataset['Fuel_Type']=encoder.fit_transform(fuel_type)
dataset.head()
```

```python
sns.kdeplot(dataset.Fuel_Type)
```

```python
#Train and Test
from sklearn.model_selection import train_test_split
x=dataset.drop(["Car_Name","Selling_Price","Kms_Driven"],axis=1)
y=dataset["Selling_Price"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=2)
x_train.shape
x_train.head()
y_train.shape
y_train.head()
```
